```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generate linear data
x_linear = np.linspace(0, 10, 50)
y_linear = 2 * x_linear + 5 + np.random.randn(50) * 2   #adding noise
```

```python
#Generate non linear data
x_nonlinear = np.linspace(0, 10, 50)
y_nonlinear = 2 * np.sin(x_nonlinear) + np.random.randn(50) #adding noise
```

```python
#plotting linear data
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(x_linear, y_linear,color='blue',label='Data with Linear Relationship')
plt.plot(x_linear, 2*x_linear + 5, color='red', label='True Relationship')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Relationship')
plt.legend()

#plotting linear data
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(x_nonlinear, y_nonlinear,color='green',label='Data with Non-Linear Relationship')
plt.plot(x_nonlinear, 2*np.sin(x_nonlinear), color='orange', label='True Non-Linear Relationship')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Non-Linear Relationship')
plt.legend()
```

```python
#correlation coefficient
corr_linear  = np.corrcoef(x_linear, y_linear)[0, 1]
corr_nonlinear = np.corrcoef(x_nonlinear, y_nonlinear)[0, 1]
```

```python
print('Correlation coefficient for linear data:', corr_linear)
print('Correlation coefficient for non-linear data:', corr_nonlinear)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_linear.reshape(-1, 1), y_linear) #x = x_linear.reshape(-1, 1) because x should be 2D array
```

```python
model.score(x_linear.reshape(-1, 1), y_linear)
```

```python
plt.plot(x_linear, model.predict(x_linear.reshape(-1, 1)), color='green', label='Best Fit')
plt.scatter(x_linear, y_linear,color='blue',label='Data with Linear Relationship')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Relationship')
plt.legend()
```

```python
#non-linear data 
model2 = LinearRegression()
model2.fit(x_nonlinear.reshape(-1, 1), y_nonlinear)
model2.score(x_nonlinear.reshape(-1, 1), y_nonlinear)
```

```python
plt.plot(x_nonlinear, model2.predict(x_nonlinear.reshape(-1, 1)), color='red', label='Best Fit')
plt.scatter(x_nonlinear, y_nonlinear,color='green',label='Data with Non-Linear Relationship')
plt.plot(x_nonlinear, 2*np.sin(x_nonlinear), color='orange', label='True Non-Linear Relationship')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Non-Linear Relationship')
plt.legend()
```
